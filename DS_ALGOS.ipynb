###Linear search

```
def linear(A,key):
  for i in range(len(A)):
     if A[i]==key:
         return i
  return -1
A=[23,14,78,98,25,54]
result=linear(A,25)
print(result)
```

####Binary search

```
def binary(A,key):
  l=0
  r=len(A)-1
  while(l<=r):
     mid=(l+r)//2
     if key==A[mid]:
        return mid
     elif key<A[mid]:
        r=mid-1
     elif key>A[mid]:
        l=mid+1
  return -1
A=[23,25,45,67,98,102]
result=binary(A,25)
print(result)
  
```

###Binary search using recursion

```
def binary_recur(A,key,l,r):
  if l>r:
    return -1
  else:
    mid=(l+r)//2
    if key==A[mid]:
       return mid
    elif key<A[mid]:
      binary_recur(A,key,l,mid-1)
    elif key>A[mid]:
       binary_recur(A,key,mid+1,r)
    
A=[23,25,45,67,98,102]
result=binary(A,25)
print(result)    
    
```

###selection sorting

```
def selection(A):
  for i in range(len(A)-1):
    position=i
    for j in range(i+1,len(A)):
      if A[position]>A[j]:
        position=j
    temp=A[i]
    A[i]=A[position]
    A[position]=temp
        
A=[24,56,78,2,13,46]
print("old",A)
selection(A)
print("sorted",A)
```

###Insertion sort

```
def Insertion(A):
  n=len(A)
  for i in range(1,n):
    cvalue=A[i]
    position=i
    while position>0 and A[position-1]>cvalue:
         A[position]=A[position-1]
         position=position-1
        
    A[position]=cvalue  
A=[24,56,78,2,13,46]
print("old",A)
Insertion(A)
print("sorted",A)
```

####Bubble sort

```
def bubble_sort(A):
  n=len(A)
  for passes in range(n-1,0,-1):
    for i in range(passes):
       if A[i]>A[i+1]:
          temp=A[i]
          A[i]=A[i+1]
          A[i+1]=temp
  
  
A=[24,56,78,2,13,46]
print("old",A)
bubble_sort(A)
print("sorted",A)  
```

###counter sorting

```
def counter(A):
  n=len(A)
  maxsize=(max(A))
  cvalue=[]
  for i in range(maxsize+1):
    cvalue.append(0)
  for i in range(n):
    cvalue[A[i]]=cvalue[A[i]]+1
  i=0
  j=0
  while i<maxsize+1:
    if cvalue[i]>0:
      A[j]=i
      j=j+1
      cvalue[i]=cvalue[i]-1
    else:
      i=i+1
        
        
A=[3,5,6,7,2,4,9]
print("old",A)
counter(A)
print("sorted",A) 
    
  
```

###Radix sort

```
def radixsort(A):
    n = len(A)
    maxelement = max(A)
    digits = len(str(maxelement))
    l = []
    bins = [l] * 10
    for i in range(digits):
        for j in range(n):
            e = int((A[j] / pow(10, i)) % 10)
            if len(bins[e]) > 0:
                bins[e].append(A[j])
            else:
                bins[e] = [A[j]]
        k = 0
        for x in range(10):
            if len(bins[x]) > 0:
                for y in range(len(bins[x])):
                    A[k] = bins[x].pop(0)
                    k = k + 1


A = [63, 250, 835, 947, 651, 28]
print('Original Array:',A)
radixsort(A)
print('Sorted Array:',A)
```

####Merge sort

```
def mergesort(A, left, right):
    if left < right:
        mid = (left + right) // 2
        mergesort(A, left, mid)
        mergesort(A, mid+1, right)
        merge(A, left, mid, right)

def merge(A, left, mid, right):
    i = left
    j = mid+1
    k = left
    B = [0] * (right+1)
    while i <= mid and j <= right:
        if A[i] < A[j]:
            B[k] = A[i]
            i = i + 1
        else:
            B[k] = A[j]
            j = j + 1
        k = k + 1

    while i <= mid:
        B[k] = A[i]
        i = i + 1
        k = k + 1

    while j <= right:
        B[k] = A[j]
        j = j + 1
        k = k + 1
    for x in range(left,right+1):
        A[x] = B[x]


A = [3, 5, 8, 9, 6, 2]
print('Original Array:',A)
mergesort(A,0,len(A)-1)
print('Sorted Array:',A)
```
